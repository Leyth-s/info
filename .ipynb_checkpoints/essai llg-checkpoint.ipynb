```python
from numpy import sin, linspace
import matplotlib.pyplot as plt


def f(x, t):
    return sin(t) * sin(x)

def euler(f, x0, t):
    n = len(t)
    x = [x0]
    for k in range(0, n-1):
        h = t[k+1] - t[k]
        p1 = f(x[k], t[k])
        x.append(x[k] + h * p1)
    return x


t = linspace(0, 50, 256)
x1 = euler(f, 1, t)
plt.plot(t, x1, label="Méthode d'Euler")
plt.title("Méthode d'Euler")
plt.legend(loc='upper left')
plt.show()
```
