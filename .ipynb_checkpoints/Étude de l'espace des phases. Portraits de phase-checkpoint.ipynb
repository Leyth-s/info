```python
import matplotlib.pyplot as plt

## IIÂ° Cas d'un oscillateur harmonique non amorti

def ohna(x0, v0, h, tmax):
    def f(x, v):   #x' = v
        return v
    def g(x, v):   #v' = -x
        return -x
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h*f(x, v)
        v = v + h*g(x, v)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    #plt.plot(T, X)
    plt.plot(X, V)
    
for i in range(20):
    ohna(i, 0, 0.001, 10)
plt.title('Portrait de phase de l\'oscillateur harmonique non amorti')
plt.savefig('/home/leyth/Bureau/im1.png', dpi=600, bbox_inches='tight')
```

```python
from numpy import arange, meshgrid
def h(x, y):
    return x**2 + y**2 + x*y
X = arange(-1, 1, 0.01)
Y = arange(-1, 1, 0.01)
X, Y = meshgrid(X, Y)
Z = h(X, Y)
plt.axis('equal')
plt.contour(X, Y, Z, [0.1, 0.4, 0.5])
plt.show()
```

```python
def ohnabis(x0, v0, h, tmax):
    def f(x, v):   #x' = v
        return v
    def g(x, v):   #v' = -x
        return -x
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h*f(x, v)
        v = v + h*g(x, v)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    E_m = (max(V)**2)/2 
    X, V = meshgrid(X, V)
    Z = f(X, V)
    plt.axis('equal')
    plt.contour(X, V, Z, [E_m])
```

```python
## IIIÂ° Cas d'un oscillateur harmonique amorti

def ohamorti(x0, v0,m, h, tmax):
    def f(x, v):   #x' = v
        return v
    def g(x, v):   #v' = -2*m*v -x
        return -2*m*v - x
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h*f(x, v)
        v = v + h*g(x, v)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    plt.plot(T, X)
    plt.title('Oscillateur harmonique amorti pour m = {}'.format(round(m, 2)))
    plt.xlabel('temps')
    plt.ylabel('trajectoire')
    plt.savefig('/home/leyth/Bureau/ohamt{}.png'.format(round(m, 2)), dpi=600, bbox_inches='tight')
    plt.cla()  #on efface le tracÃ©
    plt.plot(X, V)
    plt.title('Trajectoire de phase de l\'oscillateur harmonique amorti pour m = {}'.format(round(m, 2)))
    plt.xlabel('position')
    plt.ylabel('vitesse')
    plt.savefig('/home/leyth/Bureau/ohamtf{}.png'.format(round(m, 2)), dpi=600, bbox_inches='tight')
    plt.cla()
    
    
for i in [0.1, 2]:
    ohamorti(1, 1, i, 0.001, 20)
```

```python
## IVÂ° Cas d'un oscillateur amplifiÃ©

for i in [-0.1, -2]:
    ohamorti(1, 1, i, 0.001, 20)
```

```python
## VÂ° Oscillateur de Van der Pol

def vanderpol(x0, v0,m, a, h, tmax):
    def f(x, v, a):   #x' = v
        return v
    def g(x, v, a):   #v' = -m*(x**2 - a**2)*v - x
        return -m*(x**2 - a**2)*v - x
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h*f(x, v, a)
        v = v + h*g(x, v, a)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    plt.plot(T, X)
    plt.title('Oscillateur de Van der Pol pour m = {} et a= {}'.format(round(m, 2), a))
    plt.xlabel('temps')
    plt.ylabel('position')
    plt.savefig('/home/leyth/Bureau/vdpt{}{}.png'.format(round(m, 2),a), dpi=600, bbox_inches='tight')
    plt.cla()  #on efface le tracÃ©
    plt.plot(X, V)
    plt.title('Trajectoire de phase de l\'oscillateur de Van der Pol pour m = {} et a = {}'.format(round(m, 2), a))
    plt.xlabel('position')
    plt.ylabel('vitesse')
    plt.savefig('/home/leyth/Bureau/vdptf{}{}.png'.format(round(m, 2), a), dpi=600, bbox_inches='tight')
    plt.cla()    

vanderpol(0.5, 0, 0.1, 1, 0.01, 100)
vanderpol(0.5, 0, 2, 2, 0.01, 100)
```

```python
## VIÂ° Cas du pendule simple

from numpy import sin, pi

def pendule(x0, v0, h, tmax):
    def f(x, v):   #x' = v
        return v
    def g(x, v):   #v' = -sin(x)
        return -sin(x)
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h*f(x, v)
        v = v + h*g(x, v)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    #plt.plot(T, X)
    plt.plot(X, V, 'b', linewidth=0.3)
    plt.axis('equal')
    
    
pendule(2*pi/3, 0, 0.001, 20)
plt.grid()
plt.savefig('/home/leyth/Bureau/penduletf{}.png'.format(x0), dpi=600, bbox_inches='tight')
```

```python
for x0 in [-7 + i*0.4 for i in range(36)]:
    pendule(x0, 0, 0.001, 20)
plt.grid()
plt.savefig('/home/leyth/Bureau/pendulepf{}.png'.format(x0), dpi=600, bbox_inches='tight')
```

```python
for v0 in [0.1*i for i in range(23)]:
    pendule(0, v0, 0.001, 10)
plt.grid()
plt.savefig('/home/leyth/Bureau/pendulepf{}.png'.format(x0), dpi=600, bbox_inches='tight')
```
