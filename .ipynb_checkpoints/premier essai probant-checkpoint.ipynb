```python
from numpy import exp, linspace, pi
import matplotlib.pyplot as plt


def f(t, y):
    return -3*y

def eulerllg(f, x0, t):
    n = len(t)
    x = [x0]
    for k in range(0, n-1):
        h = t[k+1] - t[k]
        p1 = f(t[k], x[k])
        x.append(x[k] + h * p1)
    return x

def euler(f, t0, tf, y0, n):
    t, y = t0, y0
    h = (tf - t0) / n
    temps = [t0]
    fonction = [y0]
    for i in range(n):
        y = y + h*f(t, y)
        t = t + h
        temps.append(t)
        fonction.append(y)
    plt.plot(temps, fonction)
    plt.title("Méthode d'Euler - méthode 2")


t = linspace(0, 5, 100)
x1 = eulerllg(f, 1, t)
plt.plot(t, x1, label="Méthode d'Euler")
plt.title("Méthode d'Euler - programme llg")
E = [exp(-3*i) for i in t]
plt.plot(t, E, label="Solution exacte")
plt.legend(loc='upper left')
plt.show()
```
