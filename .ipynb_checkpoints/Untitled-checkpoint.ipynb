```python

##Methode d'Euler
import numpy as np
import matplotlib.pyplot as plt

#Euler seconde methode avec des listes sans indices:
#definition de f:
def f(x, y):
    return x*y
def euler(f, t0, y0, T, N):
    h = (T-t0)/N
    #initialisation a la liste vide:
    X = []
    Y = []
    #conditions initiales:
    x = t0
    y = y0
    for k in range(N):
        X.append(x)
        Y.append(y)
        y = y + h*f(x, y) #attention l'ordre entre les lignes compte
        x = x + h
    return X, Y

#application de la seconde methode:
X, Y = euler(f, 0, 1, 6, 100000)
#la solution theorique decalee de 200:
def g(x):
    return np.exp((x**2)/2) + 20000000
#trace des courbes
T = np.linspace(0, 6, 100000)
U = g(T)
plt.plot(T, U)
plt.plot(X, Y)
plt.title("solution theorique decalee de 20000000 vers le haut \n" "et solution d'Euler de l'E.D. y' = xy ")
plt.xlabel("abscisse x")
plt.ylabel("fonction y(x)")
plt.legend(["solution theorique decalee de 20000000 vers le haut", "solution d'Euler seconde methode"])

```

```python
def ohna(x0, v0, h, tmax):
    def f(x, v):   #v' = x*v
        return x*v
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h
        v = v + h*f(x, v)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    #plt.plot(T, X)
    plt.plot(X, V)
ohna(0, 1, 0.001, 2)
```
