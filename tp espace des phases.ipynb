```python
import matplotlib.pyplot as plt

## II Cas d'un oscillateur harmonique non amorti

def ohna(x0, v0, h, tmax):
    def f(t, x, v):   #x' = v
        return v
    def g(t, x, v):   #v' = -x
        return -x
    t, x, v = 0, x0, v0
    T = [0]
    X = [x0]
    V = [v0]
    n = int(tmax // h)
    for _ in range(n):
        x = x + h*f(t, x, v)
        v = v + h*g(t, x, v)
        t += h
        T.append(t)
        X.append(x)
        V.append(v)
    #plt.plot(T, X)
    plt.plot(X, V)
    
for i in range(20):
    ohna(i, 0, 0.001, 10)
plt.title('Portrait de phase de l\'oscillateur harmonique non amorti')
plt.savefig('/home/leyth/Bureau/im1.png', dpi=600, bbox_inches='tight')
```
