```python
from numpy import linspace
import matplotlib.pyplot as plt


def f(t, x, y):
    return x*(3-2*y)

def g(t, x, y):
    return y*(-4 +x)

def euler2(f, g, t0, tf, x0, y0, n):
    t, x, y = t0, x0, y0
    h = (tf - t0) / n
    temps = [t0]
    lapins = [x0]
    renards = [y0]
    for _ in range(n):
        x, y = x + h*f(t, x, y), y + h*g(t, x, y)
        t += h
        temps.append(t)
        lapins.append(x)
        renards.append(y)
    plt.plot(temps, lapins, 'b', label='lapins')
    plt.plot(temps, renards, 'g', label='renards')
    #plt.plot(lapins, renards, label='trajectoire de phase')
    plt.legend(loc='upper left')
    
```

```python
euler2(f, g, 0, 10, 5, 3, 1000)
```
